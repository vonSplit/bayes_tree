```python
import sys
import os
import numpy as np
from pathlib import Path

def find_project_root(marker_files=("CMakeLists.txt", ".git")):
    """Walk up directories until one contains any of the marker_files"""
    path = Path.cwd().resolve()
    for parent in [path] + list(path.parents):
        if any((parent / f).exists() for f in marker_files):
            return parent
    raise FileNotFoundError("Could not find project root")

# Detect project root
project_root = find_project_root()

# Path to built Python module relative to project root
build_python = project_root / "build" / "python"
#build_python = os.path.abspath(os.path.join(os.getcwd(), '..', 'build', 'python'))

print("Adding to sys.path:", build_python)
sys.path.insert(0, str(build_python))

import pybayes_tree as pbt
#print(pybayes_tree)
#from pybayes_tree import DirichletDistribution
```

```python
alpha = [2.0, 3.0, 5.0]
d = pbt.DirichletDistribution(alpha)
mean_point = d.mean()

extreme_point = [0.01, 0.01, 0.98]

logpdf_mean = d.log_pdf(mean_point)
logpdf_extreme = d.log_pdf(extreme_point)

print(f"logpdf(mean)    =  {logpdf_mean}\nlogpdf(extreme) = {logpdf_extreme}")
assert logpdf_mean > logpdf_extreme
```
